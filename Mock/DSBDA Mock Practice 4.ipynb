```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.set_theme(style="darkgrid")
```

### Count of passengers survived or not survived

```python
plt.figure(figsize=(12,12))
sns.countplot(x='survived', hue='survived', palette='Set1', data=df)
```

# Count of survival of Passengers by class

```python
plt.figure(figsize=(12,12))
sns.countplot(x='class', hue='survived', palette='Set1', data=df)
```

# Count survival of passengers by class and age

```python
plt.figure(figsize=(12,12))
sns.boxplot(x='class', y='age',hue='survived' ,palette='Set1', data=df)
```

# Count of passengers per class

```python
plt.figure(figsize=(12,12))
sns.countplot(x='class', hue="class", palette='Set1', data=df)
```

# Countplot based on Gender

```python
plt.figure(figsize=(12,12))
sns.countplot(x='class', hue='sex', palette='Set1', data=df)
```

# Survival status based of gender and age

```python
plt.figure(figsize=(12,12))
sns.boxplot(data=df, x='sex', y='age', hue='survived')
```

```python
sns.pairplot(data=df, hue='survived')
```
