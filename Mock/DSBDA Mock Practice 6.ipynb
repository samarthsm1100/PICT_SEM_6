```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
```

```python
column_names = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Class']
df = pd.read_csv(csv_url, names=column_names)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
fig, axs = plt.subplots(2,2, figsize=(20, 24))

sns.histplot(data=df, x='Sepal_Length', hue='Class', ax=axs[0,0])
sns.histplot(data=df, x='Sepal_Width', hue='Class', ax=axs[0,1])
sns.histplot(data=df, x='Petal_Length', hue='Class', ax=axs[1,0])
sns.histplot(data=df, x='Petal_Width', hue='Class', ax=axs[1,1])
```

```python
fig, axs = plt.subplots(2,2, figsize=(20, 24))

sns.boxplot(data=df, x='Sepal_Length',y='Class', hue='Class', ax=axs[0,0])
sns.boxplot(data=df, x='Sepal_Width',y='Class', hue='Class', ax=axs[0,1])
sns.boxplot(data=df, x='Petal_Length',y='Class', hue='Class', ax=axs[1,0])
sns.boxplot(data=df, x='Petal_Width',y='Class', hue='Class', ax=axs[1,1])
```

```python
fig, axs = plt.subplots(1,2,figsize=(8,4))

sns.scatterplot(data=df, x='Sepal_Length', y='Sepal_Width', hue="Class", ax=axs[0])
sns.scatterplot(data=df, x='Petal_Length', y='Petal_Width', hue="Class", ax=axs[1])
```
