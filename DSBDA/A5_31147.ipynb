### Import required libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import set_theme
import numpy as np
```

### Load dataset titanic into dataframe

```python
titanic = sns.load_dataset('titanic')
```

### Data Cleaning

```python
titanic = titanic.dropna(subset=['age'])
df = titanic
```

```python
titanic.isnull().sum()
```

```python
Q1=df['age'].quantile(0.25)
Q3=df['age'].quantile(0.75)
IQR=Q3-Q1
```

```python
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
whisker_width = 1.5
lower_whisker = Q1 -(whisker_width*IQR)
upper_whisker = Q3 + whisker_width*IQR
df['age']=np.where(df['age']>upper_whisker,upper_whisker,np.where(df['age']<lower_whisker,lower_whisker,df['age']))
```

### Boxplot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=df)
plt.title('Age distribution by Gender and Survival')
plt.show()
```

### Conclusion :

1. Out of all the passangers on Titanic, there were more female passangers survived than male passangers.

2. Most of the survived passangers are of middle age.
